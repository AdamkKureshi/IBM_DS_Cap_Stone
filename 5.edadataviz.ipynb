## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
# Group by Orbit and calculate mean success rate
orbit_success = df.groupby('Orbit')['Class'].mean().reset_index()

# Sort by success rate for cleaner visual (optional)
orbit_success = orbit_success.sort_values(by='Class', ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Orbit', y='Class', data=orbit_success)
plt.xlabel('Orbit Type', fontsize=14)
plt.ylabel('Success Rate', fontsize=14)
plt.title('Success Rate by Orbit Type', fontsize=16)
plt.xticks(rotation=45)
plt.ylim(0, 1)  # since it's a probability
plt.show()
```

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
sns.catplot(
    x="FlightNumber",
    y="Orbit",
    hue="Class",
    data=df,
    aspect=2,
    height=6
)

plt.xlabel("Flight Number", fontsize=14)
plt.ylabel("Orbit Type", fontsize=14)
plt.title("Launch Success by Orbit Type Over Flight Numbers", fontsize=16)
plt.show()
```

```python
# Group by Orbit and FlightNumber, then calculate success rate
summary_df = df.groupby(['Orbit', 'FlightNumber'])['Class'].mean().reset_index()

# Get overall success rate by Orbit
orbit_summary = df.groupby('Orbit')['Class'].agg(['count', 'mean']).reset_index()
orbit_summary.columns = ['Orbit', 'Launch_Count', 'Success_Rate']
orbit_summary = orbit_summary.sort_values(by='Success_Rate', ascending=False)

# Print summary
print("📊 Summary of Launch Success Trends by Orbit:\n")
for _, row in orbit_summary.iterrows():
    orbit = row['Orbit']
    count = int(row['Launch_Count'])
    rate = row['Success_Rate']
    
    trend = ""
    if rate == 1.0:
        trend = "💯 Perfect success record"
    elif rate > 0.7:
        trend = "✅ Generally successful"
    elif rate > 0.4:
        trend = "⚠️ Mixed success"
    else:
        trend = "❌ High failure rate"
    
    print(f"➡️ Orbit: {orbit.ljust(6)} — Launches: {count:<3} — Success Rate: {rate:.2%} — {trend}")
```

You can observe that in the LEO orbit, success seems to be related to the number of flights. Conversely, in the GTO orbit, there appears to be no relationship between flight number and success.


### TASK  5: Visualize the relationship between Payload Mass and Orbit type


Similarly, we can plot the Payload Mass vs. Orbit scatter point charts to reveal the relationship between Payload Mass and Orbit type


```python
# Plot a scatter point chart with x axis to be Payload Mass and y axis to be the Orbit, and hue to be the class value
sns.catplot(
    x="PayloadMass",
    y="Orbit",
    hue="Class",
    data=df,
    aspect=2,
    height=6
)

plt.xlabel("Payload Mass (kg)", fontsize=14)
plt.ylabel("Orbit Type", fontsize=14)
plt.title("Launch Success by Payload Mass and Orbit Type", fontsize=16)
plt.show()
```

```python
# Group by Orbit and summarize payload mass and success
payload_summary = df.groupby('Orbit').agg({
    'PayloadMass': ['count', 'mean', 'min', 'max'],
    'Class': 'mean'
}).reset_index()

# Clean up column names
payload_summary.columns = ['Orbit', 'Launch_Count', 'Avg_PayloadMass', 'Min_Payload', 'Max_Payload', 'Success_Rate']

# Sort by average payload
payload_summary = payload_summary.sort_values(by='Avg_PayloadMass', ascending=False)

# Print summary
print("📦 Summary of Payload Mass and Launch Success by Orbit:\n")
for _, row in payload_summary.iterrows():
    print(
        f"➡️ Orbit: {row['Orbit'].ljust(6)} | Launches: {row['Launch_Count']:>2} | "
        f"Payload: {row['Avg_PayloadMass']:.1f} kg avg "
        f"(min: {row['Min_Payload']:.0f}, max: {row['Max_Payload']:.0f}) | "
        f"Success Rate: {row['Success_Rate']:.2%}"
    )
```

With heavy payloads the successful landing or positive landing rate are more for Polar,LEO and ISS.

However, for GTO, it's difficult to distinguish between successful and unsuccessful landings as both outcomes are present.


### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
# Step 1: Extract Year (already done)
year = []
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year

# Replace Date column with just year
df["Date"] = Extract_year()

# Step 2: Group by Year and calculate average success rate
yearly_success = df.groupby('Date')['Class'].mean().reset_index()
yearly_success.columns = ['Year', 'SuccessRate']

# Step 3: Plot the line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_success, x='Year', y='SuccessRate', marker='o', linewidth=2)
plt.title("📈 SpaceX Yearly Launch Success Trend", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Success Rate", fontsize=14)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

```python
print("📊 Yearly Launch Success Rate Summary:\n")

for _, row in yearly_success.iterrows():
    year = row['Year']
    rate = row['SuccessRate']
    
    if rate == 1.0:
        verdict = "💯 Perfect success year"
    elif rate >= 0.8:
        verdict = "✅ High success rate"
    elif rate >= 0.5:
        verdict = "⚠️ Moderate success"
    elif rate > 0:
        verdict = "❌ Low success rate"
    else:
        verdict = "❌ No successful landings"

    print(f"➡️ {year}: Success Rate = {rate:.2%} — {verdict}")
```

you can observe that the sucess rate since 2013 kept increasing till 2020
